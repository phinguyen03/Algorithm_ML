```python
import numpy as np
```

```python
population_size=4
bit_size = 3 
def initi(population_size,bit_size):
    X_range=[]
    y_range=[]
    for i in range(population_size):
        X=np.random.randint(0,2,size=bit_size) #create binary value in array (vector with 3 value)
        y=np.random.randint(0,2,size=bit_size)
        X_range.append(X)
        y_range.append(y)
    return X_range,y_range

# fitness function
def function(x,y):
    result=[]
    for i in range(population_size):
        func=np.sum(x[i]**2)+np.sum(y[i]**2)  #sum those binary value -> decimal
        result.append(func)
    return result

# statistic choice
def statistic(result):
    statistic_val=[]
    sum_val=np.sum(result)
    for i in range(len(result)):
        choice=result[i]/sum_val
        statistic_val.append(choice)
    return statistic_val

# offspring over: lai ghep
def crossover(parent1_X, parent1_y, parent2_X, parent2_y):
    
    child1_X = np.copy(parent1_X) #copy the orginal data to exchange in the next step
    child1_y = np.copy(parent1_y)
    child2_X = np.copy(parent2_X)
    child2_y = np.copy(parent2_y)
    
    for i in range(bit_size):  #loop over the binary bit
        #uniform crossover (Uniform crossover selects the two parents for  crossover. It creates two child offspring of n genes selected from both of the parents uniformly. The random real number decides whether the first child select the ith genes from first or second parent [1])
        if np.random.random()<0.5:  #this is condition of crossover: su giao thoa in mating pool, condition have to (0,1) that 100% correct, if we set out of 0.5 example 1 then it give True value, that means everything in parents will be exchange than changing absolutely feature of parents -> it's wrong
            child1_X[i] = parent2_X[i]  #exchange bit value example x1=100 and x2=101 ->x1=101 and x2=100 
            child2_X[i] = parent1_X[i]
            child1_y[i] = parent2_y[i]
            child2_y[i] = parent1_y[i]
    
    return child1_X, child1_y, child2_X, child2_y

#second crossover
def second_crossover(parents1_X,parents1_y,parents2_X,parents2_y):
    second_child1_X=np.copy(parents1_X)
    second_child1_y=np.copy(parents1_y)
    second_child2_X=np.copy(parents2_X)
    second_child2_y=np.copy(parents2_y)
    for i in range(bit_size):
        if np.random.random() <0.5:
            second_child1_X[i]=parents2_X[i]
            second_child2_X[i]=parents1_X[i]
            second_child1_y[i]=parents2_y[i]
            second_child2_y[i]=parents1_y[i]
    return second_child1_X,second_child2_X,second_child1_y,second_child2_y

#fitness function after offspring
def fitness_offspring(x,y):
    r=[]
    for i in range(2):
        func=np.sum(x[i]**2)+np.sum(y[i]**2)
        r.append(func)
    return r

X_range,y_range=initi(population_size,bit_size)
val=function(X_range,y_range)
statistic_value=statistic(val)


# select parents for mating pool
num_parents = 2
parents_indices = np.random.choice(range(population_size), size=num_parents, replace=False, p=statistic_value) #choosing random parents value which have higher statistic
parents_X = [X_range[i] for i in parents_indices]
parents_y = [y_range[i] for i in parents_indices]

#child
childX=[]
childy=[]
child1_X, child1_y, child2_X, child2_y  = crossover(parents_X[0], parents_y[0], parents_X[1], parents_y[1])

childX.append(child1_X)
childX.append(child2_X)
childy.append(child1_y)
childy.append(child2_y)

new_val=fitness_offspring(childX,childy)

#sum,mean,max
sum=np.sum(new_val)
mean=np.mean(new_val)
max=np.max(new_val)

# the second generation
second_X=[]
second_y=[] 
for i in childX:
    second_X.append(i)
for j in childy:
    second_y.append(j)

the_second_population=2
num_parents_2=1

second_val=fitness_offspring(second_X,second_y) 


# second_statistic_value=statistic(second_val)
# second_parents_prob=np.random.choice(range(the_second_population),size=num_parents_2,replace=False,p=second_statistic_value)
# #choose random second parents
# second_parentX=[second_X[i] for i in second_parents_prob]
# second_parenty=[second_y[i] for i in second_parents_prob]

se_child1_X, se_child1_y, se_child2_X, se_child2_y  = crossover(second_X[0], second_y[0], second_X[1], second_y[1])

#list of second child
second_childX=[]
second_childy=[]

second_childX.append(se_child1_X)
second_childX.append(se_child2_X)
second_childy.append(se_child1_y)
second_childy.append(se_child2_y)

second_fitness=fitness_offspring(second_childX,second_childy)

#sum,mean,max
sum_second=np.sum(second_fitness)
mean_second=np.mean(second_fitness)
max_second=np.max(second_fitness)

print(f'X: {X_range}')
print(f'y: {y_range}') 
print(f'Fxk: {val}')
print(f'statistic_choice: {statistic_value}')

print("Parents X:", parents_X)
print("Parents y:", parents_y)
print("Child 1 X:", child1_X)
print("Child 2 X:", child2_X)
print("Child 1 y:", child1_y)
print("Child 2 y:", child2_y)

print("-"*60)
print(f"list child X: {childX}")
print(f"list child y: {childy}")

print(f"fitness function after offspring: {new_val}")
print(f"Sum value: {sum}")
print(f"Mean value: {mean}")
print(f"Max value: {max}")
print("-"*60)
print("the second generation")
print("second X:", second_X)
print("Second y:", second_y)
print(f"fxk2: {second_val}")
# print(f'second statistic_choice: {second_statistic_value}')
print("Second Parents X:", second_X)
print("Second Parents y:", second_y)

print(f"Second childX: {second_childX}")
print(f"Second childy: {second_childy}")

print(f"Second fitness function: {second_fitness}")
print(f"Sum value: {sum_second}")
print(f"Mean value: {mean_second}")
print(f"Max value: {max_second}")
```
